## Seção 2: Criação de Arrays

> Arrays são a principal estrutura de dados do NumPy.
Eles são semelhantes a listas em Python, mas permitem operações matemáticas e manipulação de dados de maneira mais eficiente.

### Criação de um array a partir de uma lista

```
array1 = np.array([1, 2, 3, 4, 5])
print("Array criado a partir de uma lista:")
print(array1)
```

## Seção 3: Operações básicas

> NumPy permite realizar operações matemáticas e lógicas de forma vetorizada, o que significa que podemos aplicar operações a todos os elementos de um array de uma só vez.
Isso torna as operações com arrays muito eficientes.

> Vamos criar os arrays que serão usados nas operações

```
array1 = np.array([1, 2, 3, 4, 5])
array2 = np.array([10, 20, 30, 40, 50])
```

````python
# tente executar o código acima aqui
````

#### Soma de arrays


````python
# tente executar o código acima aqui
````

#### Multiplicação de arrays

